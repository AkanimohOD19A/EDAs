```python
def display_missing(data):
    m_col = []
    for col in data.columns:
        if data[col].isnull().sum() > 0:
            print('{} columns missing values: {}'. format(col, data[col].isnull().sum()))
            print('\n')
            m_col.append(col)
    print(m_col)
```

```python
## Count of Missing Values Present in the Columns
display_missing(TransDf)
```

## Uniqueness

```python
def uniqueness(df):
    t_col = []
    for col in df.columns.tolist():
        print('{}: {} absence of Duplication'.format(col, df[col].is_unique),'\n')
        if df[col].is_unique == False:
            t_col.append(col)
    print('\n',t_col)

uniqueness(TransDf)
```

```python
uniqueness(NCL)
```

```python
uniqueness(CustDemo)
```

```python
uniqueness(CustAdd)
```
